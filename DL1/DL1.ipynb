```python
# Importing the libraries that are required for neural networking
import pandas
from keras.models import Sequential
from keras.layers.core import Dense, Activation

from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# load the dataset
dataset = pd.read_csv("diabetes.csv", header=None).values

# splitting the csv data
X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)

# create model, add dense layers one by one specifying activation function
my_first_nn = Sequential() 
my_first_nn.add(Dense(20, input_dim=8, activation='relu')) # input layer
my_first_nn.add(Dense(15, activation='relu'))
my_first_nn.add(Dense(8, activation='relu'))
my_first_nn.add(Dense(10, activation='relu'))
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer

# compile the model, adam gradient descent (optimized)
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])

# training the network
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,
                                     initial_epoch=0)
```

```python
# Evaluating the model
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
score = my_first_nn.evaluate(X_test, Y_test)
print("Loss: ", score[0]*100)
print("Accuracy: ", score[1]*100)
```

```python
# Task3
# load the data
dataset = pd.read_csv('breastcancer.csv')
X=dataset.iloc[:,2:31].values
Y=dataset.iloc[:,1].values
labelencoder_Y = LabelEncoder()
Y= labelencoder_Y.fit_transform(Y)

# splitting the csv data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.25, random_state=87)

#Normalize the data 
from sklearn.preprocessing import StandardScaler
sc =StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)

# create model, add dense layers one by one specifying activation function
my_first_nn = Sequential()
my_first_nn.add(Dense(150, input_dim=29, activation='relu')) # input layer
my_first_nn.add(Dense(20, activation='relu'))
my_first_nn.add(Dense(8, activation='relu'))
my_first_nn.add(Dense(10, activation='relu'))
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer

# compile the model, adam gradient descent (optimized)
my_first_nn.compile(loss='mean_squared_error', optimizer='sgd', metrics=['acc'])

# training the network
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100,initial_epoch=0)

```

```python
# Evaluating the model
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
score = my_first_nn.evaluate(X_test, Y_test)
print("Loss: ", score[0]*100)
print("Accuracy: ", score[1]*100)
```
